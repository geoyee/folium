```python
import folium

# 正确位置为天安门外金水桥
loc = [116.39123825186597,39.90681115817056]
loc = loc[::-1]
```

```python
m = folium.Map(
    location=loc,
    zoom_start=18,
    tiles="OpenStreetMap",
)
folium.Marker(loc).add_to(m)

m
```
